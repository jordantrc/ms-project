```python
# Test the image input/output and processing
import numpy as np
import cv2
import random
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from PIL import Image
from sklearn import preprocessing

%matplotlib inline
# preview the image
image = 'frames/00001.jpg'
crop_size = 112
```

```python
# preview the image, image details
img = Image.open(image)
print("height = %s, width = %s" % (img.height, img.width))
print("img shape = %s" % str(np.asarray(img).shape))
imshow(np.asarray(img))
```

```python
# test image flip
img_flipped = cv2.flip(np.asarray(img), flipCode=1)
imshow(np.asarray(img_flipped))
```

```python
# test resize
img_resize = np.array(cv2.resize(np.array(img), (171, 128)))
print("img_resize.shape = %s" % str(img_resize.shape))
imshow(np.asarray(img_resize))
```

```python
# center crop
crop_x = int((img_resize.shape[0] - crop_size) / 2)
crop_y = int((img_resize.shape[1] - crop_size) / 2)
print("crop_x = %s, crop_y = %s" % (crop_x, crop_y))
img_crop = img_resize[crop_x:crop_x + crop_size, crop_y:crop_y + crop_size, :]
imshow(img_crop)
```

```python
# rescale
img_crop = np.array(cv2.resize(np.array(img_resize), (crop_size, crop_size)))
imshow(img_crop)
```

```python
# random crop
crop_x = random.randint(0, img_resize.shape[0] - crop_size)
crop_y = random.randint(0, img_resize.shape[1] - crop_size)
print("crop_x = %s, crop_y = %s" % (crop_x, crop_y))
img_crop = img_resize[crop_x:crop_x + crop_size, crop_y:crop_y + crop_size, :]
imshow(img_crop)
```

```python
# normalization
img_crop = img_crop.astype(np.float32)
img_2d = img_crop.reshape((crop_size * crop_size, 3))
# use the sklearn scale function to get mean 0 and unit variance
img_scaled = preprocessing.scale(img_2d)
# reshape back to 3d array
img_scaled = img_scaled.reshape((crop_size, crop_size, 3))
print("img_scaled[0][0] = %s" % (img_scaled[0][0]))
imshow(img_scaled)
```
